```python
%load_ext autoreload
%autoreload 2

from tictactoe import TicTacToe
```

```python
tictactoe = TicTacToe()
player = 1

state = tictactoe.get_initial_state()

while True:
    print(state)
    valid_moves = tictactoe.get_valid_moves(state)
    print("valid_moves", [i for i in range(tictactoe.action_size) if valid_moves[i]])
    action = int(input(f"player {player}:"))

    if not valid_moves[action]:
        print("invalid action.")
        continue

    state = tictactoe.update_state(state, action, player)

    # check win or termination
    termination_val, terminated = tictactoe.get_value_and_terminated(state, action)
    if terminated:
        if termination_val == 1:
            print(f"Player {player} won!")
        elif termination_val == 0:
            print(f"Game terminated - draw.")
        break

    # switch to next player and start over
    player = tictactoe.get_opponent(player)
```
