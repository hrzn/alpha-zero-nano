```python
class TicTacToe:
    def __init__(self):
        self.row_count = 3
        self.column_count = 3
        self.action_size = self.row_count * self.column_count

    def initial_state(self):
        return np.zeros((self.row_count, self.column_count))
    
    def get_next_state(self, state: np.ndarray, action: tuple[int], player) -> np.ndarray:
        """ Provide the next state given a state, an action and a player.

        We represent action by a (row, column) tuple
        """
        row, column = action
        new_state = state.copy()
        new_state[row, column] = player
        return new_state
    
    def get_valid_moves(self, state):
        return state == 0

    def check_win(self, state, action):
        row, column = action
        player = state[row, column]
        return (
            np.all(state[row, :] == player) 
            or np.all(state[:, column] == player) 
            or np.all(np.diag(state) == player)
            or np.all(np.diag(np.fliplr(state)) == player)
        )
    
    
```
